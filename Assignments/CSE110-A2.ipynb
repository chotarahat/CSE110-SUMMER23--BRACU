```python
def calculate_season(temperature):
    # Convert Fahrenheit to Celsius
    celsius = (temperature - 32) * 0.56

    # Determine the season based on temperature
    if celsius < 20:
        season = "Winter"
    elif 20 <= celsius <= 25:
        season = "Autumn"
    elif 25 < celsius < 30:
        season = "Spring"
    else:
        season = "Summer"

    return season

# Get the temperature from the user
temperature_fahrenheit = float(input("Enter the temperature in degrees Fahrenheit: "))

# Calculate the season and temperature in Celsius
season = calculate_season(temperature_fahrenheit)
temperature_celsius = (temperature_fahrenheit - 32) * 0.56

# Print the results
print(f"The temperature is {temperature_celsius:.2f} degrees Celsius.")
print(f"The season is {season}.")
```
